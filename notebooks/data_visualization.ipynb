#### EDA (Exploratory Data Analysis) -  exploratory data analysis (EDA) is an approach of analyzing data sets to summarize their main characteristics, often using statistical graphics and other data visualization methods

```python
import pandas as pd

titanic_df = pd.read_csv('../data/titanic.csv')
```

```python
titanic_df.shape
```

```python
titanic_df.head()
```

```python
# null values

titanic_df.isnull().sum()/len(titanic_df)*100
```

```python
# filling values for Age using mean of ages

titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
```

```python
# treatment of cabin - drop the column

titanic_df.drop(columns=['Cabin'], inplace=True)
```

```python
# filling values for Embarked using mode 

titanic_df['Embarked'] = titanic_df['Embarked'].fillna((titanic_df['Embarked'].mode()).astype('str'))
```

```python
titanic_df.drop(columns=['Embarked'], inplace=True)
```

```python
titanic_df.head()
```
